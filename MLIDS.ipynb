# Import Data


```python
import pandas as pd
df1 = pd.read_csv('/CSVs/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
df2 = pd.read_csv('/CSVs/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
df3 = pd.read_csv('/CSVs/Friday-WorkingHours-Morning.pcap_ISCX.csv')
df4 = pd.read_csv('/CSVs/Monday-WorkingHours.pcap_ISCX.csv')
df5 = pd.read_csv('/CSVs/Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv')
df6 = pd.read_csv('/CSVs/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv')
df7 = pd.read_csv('/CSVs/Tuesday-WorkingHours.pcap_ISCX.csv')
df8 = pd.read_csv('/CSVs/Wednesday-workingHours.pcap_ISCX.csv')
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8])
df
```

```python
df.info()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df = df.drop(columns=['Fwd Header Length'])
df.shape
```

```python
df['Label'].unique()
```

```python
def renaming_class_label(df: pd.DataFrame):
    labels = {"Web Attack \x96 Brute Force": "Web Attack-Brute Force",
              "Web Attack \x96 XSS": "Web Attack-XSS",
              "Web Attack \x96 Sql Injection": "Web Attack-Sql Injection"}

    for old_label, new_label in labels.items():
        df.Label.replace(old_label, new_label, inplace=True)

# Renaming labels
renaming_class_label(df)
```

```python
df['Label'].value_counts()
```

```python
df = df.dropna()
df
```

```python
df.replace('Infinity', -1, inplace=True)

df[["Flow Bytes/s", "Flow Packets/s"]] = df[["Flow Bytes/s", "Flow Packets/s"]].apply(pd.to_numeric)
```

```python
import numpy as np

df.replace([np.inf, -np.inf, np.nan], -1, inplace=True)
```

```python
string_features = list(df.select_dtypes(include=['object']).columns)
string_features.remove('Label')
string_features
```

```python
attacks = df[df['Label'] != 'BENIGN']
no_attacks = 0

for number in attacks['Label'].value_counts():
  no_attacks += number
normal_traffic = df['Label'].value_counts()['BENIGN']
attack_vs_normal = pd.Series({'Normal Traffic': normal_traffic, 'Attack': no_attacks})

print('Dataset labels:')
number_of_labels = len(df['Label'].unique())

print('Number of labels:',number_of_labels)
```

```python
attack_vs_normal.plot(kind='pie')
plt.title('CICIDS2017 distribution of traffic type')
plt.axis('equal')
plt.ylabel('')
plt.savefig('/CSVs/charts/traffic_distribution_pie.png')
```
